```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
x
```

```python
y=2*x
y
```

Line Chart

```python
plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

Bar Chart

```python
plt.bar(x,y)
plt.title("Bar Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

Scatter Plot

```python
plt.scatter(x,y)
plt.title("Bar Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```
